```python
from sympy import symbols

# n: total, f: falsy, q: quorum
n, f, q = symbols('n f q')
```

## Safety

Any two quorums must contain at least one common replica.

```python
safety = 2 * q - n > f
```

## Liveness

Correct replicas can form a quorum when all falsy replicas do not respond.

```python
liveness = n - f >= q
```

## Solve

```python
from sympy import reduce_inequalities

print(reduce_inequalities([safety, liveness], [q]))
```

To allow `q` exist:

```python
q_exists = f/2 + n/2 < -f + n
print(reduce_inequalities([q_exists], [n]))
```
